```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer
```

```python
column_title = np.append(cancer['feature_names'], ['target'])
df_cancer = pd.DataFrame(  np.c_[ cancer['data'], cancer['target'] ], columns = column_title )
df_cancer


```

```python
sns.scatterplot(x='mean radius', y='mean area', hue= 'target', data= df_cancer)
```
